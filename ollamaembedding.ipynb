Llama 3.1 embedding

```python
#ollama supports embedding models, making it possible to build retrieval augumented generation(RAG) applications that combine text prompts with existing documents or other data
```

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(OllamaEmbeddings(model="llama3.1"))
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet",
        "Beta is the second letter of the Greek alphabet"
        
    ]
    
)
```

```python
len(r1[0])
```

```python
embeddings.embed_query("What is the second letter of the greek alphabet")
```
